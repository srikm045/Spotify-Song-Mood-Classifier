```python
def spiral(matrix):
    s = []
    top, bot = 0, matrix.shape[0] - 1
    left, right = 0, matrix.shape[1] - 1

    while top <= bot and left <= right:
        for i in range(left, right + 1):
            s.append(int(matrix[top, i]))
        top += 1

        for i in range(top, bot + 1):
            s.append(int(matrix[i, right]))
        right -= 1

        if top <= bot:
            for i in range(right, left - 1, -1):
                s.append(int(matrix[bot, i]))
            bot -= 1

        if left <= right:
            for i in range(bot, top - 1, -1):
                s.append(int(matrix[i, left]))
            left += 1

    return s

print(spiral(a))
```
